```python
# Import required libraries

# !pip install PyPDF2
# !pip install textract
import PyPDF2
import textract
import re
import string
import pandas as pd
from pdfminer.high_level import extract_text
```

```python
# Scan resume & count words based on key job skills + skillsets

# Open pdf file
directory = 'drive/MyDrive/Colab Notebooks/practice/'
filename = 'Ameena-Ruffin-Resume_Material.pdf'
text = extract_text(directory + filename, 'rb')

# Convert all strings to lowercase
text = text.lower()
# Remove numbers
text = re.sub(r'\d+','', text)

# Remove punctuation
text = text.translate(str.maketrans('','',string.punctuation))

# Create dictionary with industrial and system engineering key terms by area
terms = {'Research':['quantitative','psychology', 'social', 'media','design'],      
        'Operations management':['team'],
        'Supply chain':[],
        'Project management':[],
        'Data analytics':['analytics','api','aws','big data','business intelligence','clustering','code',
                          'coding','data','database','data mining','data science','deep learning','hadoop',
                          'hypothesis test','iot','internet','machine learning','modeling','nosql','nlp',
                          'predictive','programming','python','r','sql','tableau','text mining',
                          'visualuzation'],
        'unfavorable':[]}

# Initializie score counters for each area
research = 0
operations = 0
supplychain = 0
project = 0
data = 0
unfavorable = 0

# Create an empty list where the scores will be stored
scores = []

# Obtain the scores for each area
for area in terms.keys():
        
    if area == 'Research':
        for word in terms[area]:
            if word in text:
                research +=1
        scores.append(research)
        
    elif area == 'Operations management':
        for word in terms[area]:
            if word in text:
                operations +=1
        scores.append(operations)
        
    elif area == 'Supply chain':
        for word in terms[area]:
            if word in text:
                supplychain +=1
        scores.append(supplychain)
        
    elif area == 'Project management':
        for word in terms[area]:
            if word in text:
                project +=1
        scores.append(project)
        
    elif area == 'Data analytics':
        for word in terms[area]:
            if word in text:
                data +=1
        scores.append(data)
        
    else:
        for word in terms[area]:
            if word in text:
                unfavorable +=1
        scores.append(unfavorable)
        
# Create a data frame with the scores summary
summary = pd.DataFrame(scores,index=terms.keys(),columns=['score']).sort_values(by='score',ascending=False)
summary
```

```python
#count word occurences

def word_count(str):
    counts = dict()
    words = str.split()

    for word in words:
        if word in counts:
            counts[word] += 1
        else:
            counts[word] = 1
    
    # Sort dict by values
    counts = {k: v for k, v in sorted(counts.items(), key=lambda x: x[1], reverse=True)}
    # Filter dict for values > #
    counts = {key:val for key, val in counts.items() if val > 5}

    return counts

directory = 'drive/MyDrive/Colab Notebooks/practice/'
filename = 'compiled resumes - research associate 20 hired.pdf'
from pdfminer.high_level import extract_text
text = extract_text(directory + filename,'rb')

# Convert all strings to lowercase
text = text.lower()

# Remove numbers
text = re.sub(r'\d+','',text)

# Remove punctuation
text = text.translate(str.maketrans('','',string.punctuation))

# Remove common words
for r in (("and", ""), ("", ""), ("", ""), ("the", ""), ("", ""), ("for", ""), ("", ""), ("with", ""), ("university", ""),
          ("•", ""), ("●", ""), ("present", ""), ("skills", "")):
    text = text.replace(*r)

totals = word_count(text) 

for key, value in totals.items():
    print(key, ' : ', value)
```
